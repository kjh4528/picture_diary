### 퍼셉트론
- 2차원 데이터가 기본이다. (스칼라, 벡터, 매트릭스, 텐서), 차원이 있는 데이터를 기본으로 다룬다.

```python
# 입력: 시험 성적, 출석률
x = np.array([85,90])
w = np.array([0.7, 0.3])
b = -50
z = np.dot(x,w) + b
y = 1 if z > 0 else 0
```

```python
print('가중합: ', z)
print('출력 y: ', y)
```
