```python
# StandardScaler : 평균 0, 표준편차 1 -> (값 - 평균) / 표준편차 
# MinMaxScler : 0 ~ 1 사이의 값 (양수만 존재)
# RobustScaler : 중앙값과 IQR을 사용, 이상치에 영향을 덜 받지만 연산량 많음
# 음수 표현이 의미가 있을 경우 StandardScaler 사용 등, 스케일링시 도메인 지식 필요

from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
scaler = StandardScaler()
```

```python
age_df = pd.DataFrame({'Age':[20,30,40]})
age_scaled = scaler.fit_transform(age_df[['Age']]) # 데이터프레임 형식으로 전달 
result_df = pd.DataFrame({'Age_original':age_df['Age'],'Age_scaled':age_scaled.flatten().round(2)})
result_df
```

```python
scaler = MinMaxScaler()
fare_df = pd.DataFrame({'Age':[10,60,110]})
fare_scaled = scaler.fit_transform(fare_df[['Age']])
result_df = pd.DataFrame({'Fare_original':fare_df['Age'],'Fare_scaled':fare_scaled.flatten().round(2)})
result_df
```

```python
robust_scaler = RobustScaler()
standard_scaler = StandardScaler()
fare_df = pd.DataFrame({'fare':[7.25,7.92,53.10,71.83,512.00]})

standard_values = standard_scaler.fit_transform(fare_df[['fare']])
robust_values = robust_scaler.fit_transform(fare_df[['fare']])
result_df = pd.DataFrame({'standard_values':standard_values.flatten().round(2),'robust_values':robust_values.flatten().round(2)})
result_df
```
